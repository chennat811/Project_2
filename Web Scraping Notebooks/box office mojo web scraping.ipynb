```python
from bs4 import BeautifulSoup
import requests

from datetime import datetime
import numpy as np
import pandas as pd
import re
import pickle as pkl
```

```python
#There are 21 pages on Box Office Mojo for listing box office number for documentaries
#Write a function to create the links for the 20 pages
links =[]
links.append('https://www.boxofficemojo.com/genre/sg2245914881/')
for i in range(0, 20):
    link = 'https://www.boxofficemojo.com/genre/sg2245914881/' + '?offset=' + str(i+1) +'00'
    links.append(link)
```

```python
#Obtain the source code of each link in the list of links
#Find the table in the html code and iterate through the rows, zoom into table to find row data
# Then append the information to the final list
final_list = []
for link in links:
    source_code = requests.get(link)
    soup = BeautifulSoup(source_code.text, 'html5lib')
    table = soup.find('table')
    table_rows = table.find_all('tr')
    for tr in table_rows:
        td = tr.find_all('td')
        row = [i.text for i in td]
        final_list.append(row)
final_list
```

```python
#Make a list of column names and then create a dataframe
names = ['Mojo Rank', 'Title', 'Lifetime Gross','Max Theaters', 'Opening', 'Opening Th', 'Release Date', 'Distributor']
df = pd.DataFrame(final_list, columns = names) 
final = df[1:]
```

```python
final.to_pickle('mojo.pkl')
```

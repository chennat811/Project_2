```python
"""
Function for the measure of absolute error
between actual versus predicted values (paired observation).
"""
def mae(y_true, y_pred):
    return np.mean(np.abs(y_pred - y_true))
```

```python
"""
The standard deviation of the residuals. 
It measures how far from the regression line the data points are.
"""
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

Only 1 train and test dataset since datset is small.

```python
#Split the data into train/val/test

X = df.loc[:, ['Lifetime_Gross_MOJO', 'Max_Theaters', 'Opening_Box',
                   'Opening_Th', 'User_Rating_IMDB','Number_of_votes_IMDB','CBS Films',
                   'Focus Features', 'IDP Distribution', 'IMAX', 'Lionsgate',
                   'National Geographic Entertainment', 'Neon', 'Other', 'Overture Films',
                   'Paramount Classics', 'Paramount Pictures',
                   'Sony Pictures Entertainment (SPE)', 'United Artists',
                   'Walt Disney Studios Motion Pictures', 'Warner Bros.',
                   'Warner Independent Pictures (WIP)','log_Lifetime_Gross_MOJO','log_Max_Theaters',
                   'log_Opening_Box', 'log_Opening_Th',
                   'log_User_Rating_IMDB', 'log_Number_of_votes_IMDB']]
y= df.loc[:, ['Metacritic']]
```

## Polynomial Features

```python
#Transform all features into polynomials, not including log transformed features.
X = df.loc[:, ['Lifetime_Gross_MOJO', 'Max_Theaters', 'Opening_Box',
                   'Opening_Th', 'User_Rating_IMDB','Number_of_votes_IMDB','CBS Films',
                   'Focus Features', 'IDP Distribution', 'IMAX', 'Lionsgate',
                   'National Geographic Entertainment', 'Neon', 'Other', 'Overture Films',
                   'Paramount Classics', 'Paramount Pictures',
                   'Sony Pictures Entertainment (SPE)', 'United Artists',
                   'Walt Disney Studios Motion Pictures', 'Warner Bros.',
                   'Warner Independent Pictures (WIP)']]
y = df['Metacritic']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=42)
```

```python
#transform all features into polynomials
poly_2 = PolynomialFeatures(degree=2, interaction_only = True)
poly_X_tr = poly_2.fit_transform(X_train)
poly_X_te = poly_2.fit_transform(X_test)
```

```python
#Then, transform using standard scaler
sp_X_tr = s.fit_transform(poly_X_tr)
sp_X_te = s.fit_transform(poly_X_te)
```

```python
#Plot a linear regression model using the transformed features
lr2 = LinearRegression()
lr2.fit(sp_X_tr, y_train)
lr2.score(sp_X_tr, y_train)
```

The R<sup>2</sup> score improved compared to the linear regression model with log transformed features(best model).

```python
#make a prediction
prediction2 = lr2.predict(sp_X_te)
```

```python
r2 = r2_score(y_test, prediction2)
r2_score(y_test, prediction2)
print(r2)
```

The R<sup>2</sup> score dramatically worsened, may be overfitting, run Lasso and Ridge on the model.

```python
rmse(prediction2, y_test)
```
